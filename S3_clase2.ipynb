### Teorema de Chebyshev

**Teorema.** Dado $k\geq 1$ y un conjunto de $n$ datos, al menos $(1-\frac{1}{k^2})$ de los datos estarán dentro de $k$ desviaciones estándar de su media.

Consecuencias:

* Al menos 75% de los datos están en el intervalo $(\mu - 2\sigma,\mu + 2\sigma)$.
* Al menos 88.8% de los datos están en el intervalo $(\mu - 3\sigma,\mu + 3\sigma)$

```python
contam['dioxido_nitrogeno'].mean()-3*contam['dioxido_nitrogeno'].std()
```

```python
contam['dioxido_nitrogeno'].mean()+3*contam['dioxido_nitrogeno'].std()
```

### Regla empírica

La **Regla empírica** es una regla para describir la variabilidad de un conjunto de datos que “se apilan” en la conocida forma de montículo.

**Regla empírica.** Dada una distribución de datos que tiene forma aproximada de montículo:

* El intervalo $(\mu - \sigma,\mu + \sigma)$ contiene aproximadamente 68% de los datos.

* El intervalo $(\mu - 2\sigma,\mu + 2\sigma)$ contiene aproximadamente 95% de los datos.

* El intervalo $(\mu - 3\sigma,\mu + 3\sigma)$ contiene aproximadamente 99.7% de los datos.

# Correlación

## El coeficiente de correlación de Pearson

El coeficiente de correlación de **Pearson** es una prueba que mide la relación estadística entre dos variables continuas. 

 * El coeficiente de correlación puede tomar un rango de valores de +1 a -1. 

* Un valor cercano a 0 indica que no hay asociación entre las dos variables. 

* Un valor cercano a 1 indica una asociación positiva. Es decir, a medida que aumenta el valor de una variable, también lo hace el valor de la otra. 

* Un valor cercano a -1  indica una asociación negativa; es decir, a medida que aumenta el valor de una variable, el valor de la otra disminuye.

Suposiciones vinculadas al uso del coeficiente de correlación de Pearson:

1. *Nivel de medición de las variables:* las dos variables deben ser de intervalo o de razón, aunque no es necesario que ambas tengan el mismo nivel de medición.


2. *Datos pareados:* para que el cálculo de esta medida pueda realizarse, se necesitará que los casos en cuestión tengan datos en cada variable. Si hay valores perdidos, estos registros se descartarán por completo del análisis.


3. *Normalidad bivariada:* contrario a lo que sucede en muchas investigaciones en las que se comprueba únicamente la normalidad marginal de X y Y, el uso apropiado del coeficiente de correlación de Pearson exige que se satisfaga el supuesto de normalidad bivariada; esto es, que la distribución de probabilidad conjunta de X y Y sea normal.


### Prueba de normalidad de Shapiro-Wilk


La prueba de Shapiro–Wilk se usa para contrastar la normalidad de un conjunto de datos. Se plantea como hipótesis nula que una muestra proviene de una población normalmente distribuida. Se considera uno de las pruebas más potentes para el contraste de normalidad. 


Al ser una prueba de significación estadística, se establecen dos 
hipótesis que se deberán contrastar: 

* **Hipótesis nula ($H_0$):** la muestra procede de una distribución normal. Aceptamos la $H_0$ siempre y cuando el $p$-valor de esta prueba estadística  sea mayor que 0.05. 

* **Hipótesis alternativa ($H_1$):** por el contrario, los datos no se distribuyen según un modelo de probabilidad normal. Aceptamos la $H_1$ cuando el $p$-valor de la prueba sea menor que 0.05. 

```python
from scipy.stats import shapiro

estadistico, p_value = shapiro(contam['ozono'])
print('Valor del estadístico = %.3f , p_value= %.10f' % (estadistico,p_value)) 
print('\n')

estadistico, p_value = shapiro(contam['dioxido_azufre'])
print('Valor del estadístico = %.3f , p_value= %.10f' % (estadistico,p_value)) 
print('\n')

estadistico, p_value = shapiro(contam['dioxido_nitrogeno'])
print('Valor del estadístico = %.3f , p_value= %.10f' % (estadistico,p_value)) 
print('\n')

estadistico, p_value = shapiro(contam['monoxido_carbono'])
print('Valor del estadístico = %.3f , p_value= %.10f' % (estadistico,p_value)) 
print('\n')

estadistico, p_value = shapiro(contam['PM10'])
print('Valor del estadístico = %.3f , p_value= %.10f' % (estadistico,p_value)) 
print('\n')
```

## El coeficiente de correlación de Spearman

El coeficiente de correlación de Spearman ($ρ$), es una medida de la correlación entre dos variables aleatorias. Esta es una medida no paramétrica de la correlación de rango. Se utiliza principalmente para el análisis de datos para medir la relación entre dos variables.

Una gran ventaja de la prueba de correlación de Spearman es que no importa si las variables tiene una distribución normal, y poseé las propiedades siguiente:

* **Rangos de datos:** En lugar de trabajar con los valores originales de los datos, la correlación de Spearman se calcula a partir de los rangos de los datos. 

* **Cálculo de correlación:** Su coeficiente ρ varía entre 1 y -1, el 1 indica que es perfecta, están sincronizadas, y el -1 indica que una correlación negativa perfecta, ie, mientras uno sube el otro baja y el 0 indica que no hay correlación.

* **Monotonía y no linealidad:** es útil para detectar si habrá relaciones monotónicas
	
* **Robustez:** esto implica que ese índice es menos sensible a datos atípicos.


```python
correlacion=contam.corr(method='spearman')
```
