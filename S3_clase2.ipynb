### Teorema de Chebyshev

Consecuencias:

* Al menos 75% de los datos estÃ¡n en el intervalo $(\mu - 2\sigma,\mu + 2\sigma)$.
* Al menos 88.8% de los datos estÃ¡n en el intervalo $(\mu - 3\sigma,\mu + 3\sigma)$

# CorrelaciÃ³n

## El coeficiente de correlaciÃ³n de Pearson

El coeficiente de correlaciÃ³n de **Pearson** es una prueba que mide la relaciÃ³n estadÃ­stica entre dos variables continuas. 

 * El coeficiente de correlaciÃ³n puede tomar un rango de valores de +1 a -1. 

* Un valor cercano a 0 indica que no hay asociaciÃ³n entre las dos variables. 

* Un valor cercano a 1 indica una asociaciÃ³n positiva. Es decir, a medida que aumenta el valor de una variable, tambiÃ©n lo hace el valor de la otra. 

* Un valor cercano a -1  indica una asociaciÃ³n negativa; es decir, a medida que aumenta el valor de una variable, el valor de la otra disminuye.

Suposiciones vinculadas al uso del coeficiente de correlaciÃ³n de Pearson:

1. *Nivel de mediciÃ³n de las variables:* las dos variables deben ser de intervalo o de razÃ³n, aunque no es necesario que ambas tengan el mismo nivel de mediciÃ³n.


2. *Datos pareados:* para que el cÃ¡lculo de esta medida pueda realizarse, se necesitarÃ¡ que los casos en cuestiÃ³n tengan datos en cada variable. Si hay valores perdidos, estos registros se descartarÃ¡n por completo del anÃ¡lisis.


3. *Normalidad bivariada:* contrario a lo que sucede en muchas investigaciones en las que se comprueba Ãºnicamente la normalidad marginal de X y Y, el uso apropiado del coeficiente de correlaciÃ³n de Pearson exige que se satisfaga el supuesto de normalidad bivariada; esto es, que la distribuciÃ³n de probabilidad conjunta de X y Y sea normal.


### Prueba de normalidad de Shapiro-Wilk


La prueba de Shapiroâ€“Wilk se usa para contrastar la normalidad de un conjunto de datos. Se plantea como hipÃ³tesis nula que una muestra proviene de una poblaciÃ³n normalmente distribuida. Se considera uno de las pruebas mÃ¡s potentes para el contraste de normalidad. 


Al ser una prueba de significaciÃ³n estadÃ­stica, se establecen dos 
hipÃ³tesis que se deberÃ¡n contrastar: 

* **HipÃ³tesis nula ($H_0$):** la muestra procede de una distribuciÃ³n normal. Aceptamos la $H_0$ siempre y cuando el $p$-valor de esta prueba estadÃ­stica  sea mayor que 0.05. 

* **HipÃ³tesis alternativa ($H_1$):** por el contrario, los datos no se distribuyen segÃºn un modelo de probabilidad normal. Aceptamos la $H_1$ cuando el $p$-valor de la prueba sea menor que 0.05. 

```python
from scipy.stats import shapiro

estadistico, p_value = shapiro(contam['ozono'])
print('Valor del estadÃ­stico = %.3f , p_value= %.10f' % (estadistico,p_value)) 
print('\n')

estadistico, p_value = shapiro(contam['dioxido_azufre'])
print('Valor del estadÃ­stico = %.3f , p_value= %.10f' % (estadistico,p_value)) 
print('\n')

estadistico, p_value = shapiro(contam['dioxido_nitrogeno'])
print('Valor del estadÃ­stico = %.3f , p_value= %.10f' % (estadistico,p_value)) 
print('\n')

estadistico, p_value = shapiro(contam['monoxido_carbono'])
print('Valor del estadÃ­stico = %.3f , p_value= %.10f' % (estadistico,p_value)) 
print('\n')

estadistico, p_value = shapiro(contam['PM10'])
print('Valor del estadÃ­stico = %.3f , p_value= %.10f' % (estadistico,p_value)) 
print('\n')
```

## El coeficiente de correlaciÃ³n de Spearman

El coeficiente de correlaciÃ³n de Spearman ($Ï$), es una medida de la correlaciÃ³n entre dos variables aleatorias. Esta es una medida no paramÃ©trica de la correlaciÃ³n de rango. Se utiliza principalmente para el anÃ¡lisis de datos para medir la relaciÃ³n entre dos variables.

Una gran ventaja de la prueba de correlaciÃ³n de Spearman es que no importa si las variables tiene una distribuciÃ³n normal, y poseÃ© las propiedades siguiente:

* **Rangos de datos:** En lugar de trabajar con los valores originales de los datos, la correlaciÃ³n de Spearman se calcula a partir de los rangos de los datos. 

* **CÃ¡lculo de correlaciÃ³n:** Su coeficiente Ï varÃ­a entre 1 y -1, el 1 indica que es perfecta, estÃ¡n sincronizadas, y el -1 indica que una correlaciÃ³n negativa perfecta, ie, mientras uno sube el otro baja y el 0 indica que no hay correlaciÃ³n.

* **MonotonÃ­a y no linealidad:** es Ãºtil para detectar si habrÃ¡ relaciones monotÃ³nicas
	
* **Robustez:** esto implica que ese Ã­ndice es menos sensible a datos atÃ­picos.
