### Listas

Una lista es una secuencia de valores. Es un tipo de contenedor usado para almacenar un número indefinido de datos arbitrarios del mismo tipo o de tipos distintos.

Una lista homogénea contiene elementos del mismo tipo.

```python
lista1 = [3,4,5] #listas
```

```python
lista2 = ['uno', 'dos', 'tres']
```

Una lista heterogénea combina diferentes tipos de archivos. 

```python
lista3 = ['uno', 2, 3.14, [10, 20, 30]]
```

Las listas en Python tienen las siguientes propiedades:

* Son ordenadas: cada elemento dentro de la lista está indexado y mantiene el orden definido.
    
* Permiten elementos duplicados: las listas pueden tener elementos con el mismo valor.

* Las listas pueden anidar, es decir, contener una lista dentro de la otra.

* Son mutables: podemos editar, eliminar o agregar elementos en una lista existente.


```python
lista1
```

```python
lista1[2]
```

```python
lista1[2] = 10 # Las listas son mutables. Esto significa que el objeto o los valores que contiene se pueden modificar
```

```python
lista1
```

```python
# añadir elementos
lista1.append(11)
```

```python
lista1
```

```python
# eliminar elementos
lista1.remove(7)
```

```python
lista1
```

## Módulos

Un módulo es un archivo de Python cuyos objetos (funciones, clases, etc.) pueden ser accedidos desde otro archivo. Se trata simplemente de una forma de organizar grandes códigos.

```python
import aritmetica

aritmetica.sumar(89, 5)
```

```python
# Una sintaxis alternativa para importar objetos desde un módulo es la siguiente.

from aritmetica import sumar

sumar(89, 5)
```

```python
# Podemos importar varios objetos separándolos por comas.

from aritmetica import sumar, restar, mult, div

print(sumar(7, 5))
print(restar(7, 5))
print(mult(7, 5))
```

```python
div(8, 2)
```

```python
# O bien, para importar todos los objetos dentro de un módulo:

from aritmetica import *
```

## Fechas y horas

El módulo **datetime** integrado de Python ofrece los tipos datetime, date y time. El tipo datetime combina la información almacenada en date y time y es el más utilizado:


```python
from datetime import datetime, date, time
```

```python
dt = datetime(2011, 10, 29, 20, 30, 21)
```

```python
dt.day
```

```python
dt.minute
```

```python
dt.date()
```

```python
dt.time()
```

El método **strftime** formatea un datetime como cadena de texto:

```python
dt.strftime('%Y-%m-%d %H:%M')
```

Las cadenas de texto se pueden convertir en objetos datetime con la función **strptime**:

```python
datetime.strptime('20091031', '%Y%m%d')
```

Cuando estamos agregando o agrupando de otro modo datos de series temporales, de vez en cuando es útil reemplazar campos de hora de una serie de datetimes (por ejemplo, sustituyendo los campos minute y second por cero):

```python
dt_hour = dt.replace(minute=0, second=1)
```

```python
dt_hour
```

```python
dt
```

```python
dt_hour = dt.replace(hour=11,minute=4, second=3)
```

```python
dt_hour
```

La diferencia de dos objetos datetime produce un tipo **datetime.timedelta**:

```python
dt2 = datetime(2011, 11, 15, 22, 30)
```

```python
dt2
```

```python
dt
```

```python
delta = dt2-dt
```

```python
delta
```

```python
type(delta)
```

El resultado timedelta(17, 7179) indica que el timedelta codifica un desplazamiento de 17 días y 7,179 segundos.

Sumar un timedelta a un datetime produce un nuevo datetime movido de su sitio:

```python
dt
```

```python
dt+delta
```
