### Problema 1

```python
def estan(L):
    P=[]
    for i in L:
        P.append((i-min(L))/(max(L)-min(L)))
    return P
    
```

```python
L=[42,45,2,3,5,456,445,90]
```

```python
estan(L)
```

### Problema 5

```python
contam = pd.read_excel('contam2020.xlsx')
```

```python
contam.columns = ['Fecha','ozono','dioxido_azufre',
                  'dioxido_nitrogeno','monoxido_carbono',
                  'PM10']
contam = contam.set_index("Fecha")
contam
```

```python
contam.plot(subplots = True)  
plt.show()
```

```python
contam.boxplot(rot=45,grid=False)
```

```python
from scipy.stats import shapiro

estadistico, p_value = shapiro(contam['ozono'])
print('Valor del estadístico = %.3f , p_value= %.10f' % (estadistico,p_value)) 
print('\n')

estadistico, p_value = shapiro(contam['dioxido_azufre'])
print('Valor del estadístico = %.3f , p_value= %.10f' % (estadistico,p_value)) 
print('\n')

estadistico, p_value = shapiro(contam['dioxido_nitrogeno'])
print('Valor del estadístico = %.3f , p_value= %.10f' % (estadistico,p_value)) 
print('\n')

estadistico, p_value = shapiro(contam['monoxido_carbono'])
print('Valor del estadístico = %.3f , p_value= %.10f' % (estadistico,p_value)) 
print('\n')

estadistico, p_value = shapiro(contam['PM10'])
print('Valor del estadístico = %.3f , p_value= %.10f' % (estadistico,p_value)) 
print('\n')
```

```python
correlacion=contam.corr(method='spearman')
correlacion
```
